```python
import pandas as pd

data = pd.read_csv('ground_truth_annotation_file.csv')
data.head(5)
```

```python
data.info()
```

```python
import re

data['medical_category'] = data['file'].apply(lambda x: re.findall(r'\d+_([a-zA-z ]+).txt', x)[0] if re.findall(r'\d+_([a-zA-z ]+).txt', x) else x)
```

```python
data['medical_category'].value_counts()
```

```python
data.groupby('medical_category')['class'].value_counts()
```

```python
data['class'].value_counts()
```

```python
data.groupby('file')[['text','class']].value_counts()
```

```python
s=data.groupby('file')[['text','class']].value_counts()
```

```python
txts = data['file'].unique().tolist()


principal_words = []
for txt in txts:
    principal_words.append( data.loc[(data['file'] == txt) & (data['class']=='Disease_Syndrome') ]['text'].values)

new_data = pd.DataFrame({
    'file': txts,
    'principal_words': principal_words
})
    
new_data.head(5)
```
